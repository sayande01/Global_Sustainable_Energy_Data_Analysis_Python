# Electricity access vs GDP Per capita

```python
plt.figure(figsize=(8,6))
sns.regplot(x='GDP_Per_Capita($)', y='Electricity_Access(%)', data=df)

# Set the labels and title
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Electricity Access (%)')
plt.title('Electricity Access(% of total population) vs. GDP Per Capita')

# Show the plot
plt.show()
```

# Renewable_Energy_Share(%) vs. GDP_Growth(%)

```python
plt.figure(figsize=(8,6))
sns.regplot(x='GDP_Growth(%)', y='Renewable_Energy_Share(%)', data=df)

# Set the labels and title
plt.xlabel('GDP_Growth(%)')
plt.ylabel('Renewable_Energy_Share(%)')
plt.title('Renewable_Energy_Share(%) vs. GDP_Growth(%)')

# Show the plot
plt.show()
```

# Electricity Access(% of total population) vs. Renewable_Energy_Share(%

```python
plt.figure(figsize=(10,6))
sns.regplot(x='Renewable_Energy_Share(%)', y='Electricity_Access(%)', data=df)

# Set the labels and title
plt.xlabel('Renewable_Energy_Share(%)')
plt.ylabel('Electricity Access (%)')
plt.title('Electricity Access(% of total population) vs. Renewable_Energy_Share(%)')

# Show the plot
plt.show()
```

# CO2 Emission over the years

```python
plt.figure(figsize=(12,6))
ax = sns.lineplot(x = 'Year', y = 'CO2_Emissions_kt', data = df)
ax.ticklabel_format(style='plain', axis='y')
plt.xticks(range(2000, 2021))
plt.show()
```

# Countries with most Land Area

```python
df1 = df.groupby('Country')['Land_Area(Km2)'].mean().sort_values(ascending = False).head(10)
df1
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x = df1.index, y = df1.values, palette = 'dark')
ax.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
# Visualize histograms for each numerical column
df.hist(figsize=(12, 10))
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
# Display the correlation matrix
print(correlation_matrix)
```

```python
plt.figure(figsize=(14, 12))
# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5,
            cbar_kws={"shrink": .75}, xticklabels=correlation_matrix.columns, yticklabels=correlation_matrix.columns,
            annot_kws={"size": 8}, square=True)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=45, ha='right')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
